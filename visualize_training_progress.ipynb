```python
import pandas as pd
import matplotlib.pyplot as plt
import glob

# Load all loss logs
log_files = glob.glob('logs/gpt2_medium_*.csv')
logs = []
for f in log_files:
    df = pd.read_csv(f)
    logs.append(df)
loss_df = pd.concat(logs).sort_values('step').reset_index(drop=True)

plt.figure(figsize=(10,5))
plt.plot(loss_df['step'], loss_df['loss'], label='Training Loss')
plt.xlabel('Step')
plt.ylabel('Loss')
plt.title('GPT-2 Training Loss Curve')
plt.grid(True)
plt.legend()
plt.show()
```
